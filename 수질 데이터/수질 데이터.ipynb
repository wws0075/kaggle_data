# 데이터 정보

- pH value
    - PH는 물의 산-염기 균형을 평가하는 중요한 매개변수입니다. 또한 물 상태의 산성 또는 알칼리성 상태를 나타내는 지표입니다. WHO는 pH의 최대 허용 한계를 6.5에서 8.5로 권장했습니다. 현재 조사 범위는 6.52~6.83으로 WHO 기준 범위에 있다. <br/>
- 경도(Hardness)
    - 경도는 주로 칼슘과 마그네슘 염에 의해 발생합니다. 이 염은 물이 이동하는 지질 퇴적물에서 용해됩니다. 물이 경도 생성 물질과 접촉하는 시간은 원수의 경도를 결정하는 데 도움이 됩니다. 경도는 원래 칼슘과 마그네슘에 의해 비누를 침전시키는 물의 능력으로 정의되었습니다.
- 고형물(총 용존 고형물 - TDS) / Solids (Total dissolved solids - TDS)
    - 물은 칼륨, 칼슘, 나트륨, 중탄산염, 염화물, 마그네슘, 황산염 등과 같은 광범위한 무기 및 일부 유기 미네랄 또는 염을 용해하는 능력이 있습니다. 이러한 미네랄은 원하지 않는 맛과 물의 외관에 희석된 색상을 생성합니다. 이것은 물 사용의 중요한 매개변수입니다. TDS 값이 높은 물은 물이 고도로 광물화되었음을 나타냅니다. TDS의 바람직한 한도는 500 mg/l이고 최대 한도는 음용으로 규정된 1000 mg/l입니다.
- 클로라민(Chloramines)
    - 염소와 클로라민은 공공 상수도 시스템에서 사용되는 주요 소독제입니다. 클로라민은 암모니아가 염소에 첨가되어 음용수를 처리할 때 가장 일반적으로 형성됩니다. 리터당 최대 4밀리그램(mg/L 또는 4ppm(4ppm))의 염소 수준은 식수에서 안전한 것으로 간주됩니다.
- 황산염(Sulfate)
    - 황산염은 광물, 토양 및 암석에서 발견되는 자연 발생 물질입니다. 그들은 주변 공기, 지하수, 식물 및 음식에 존재합니다. 황산염의 주요 상업적 용도는 화학 산업입니다. 해수의 황산염 농도는 리터당 약 2,700밀리그램(mg/L)입니다. 일부 지리적 위치에서는 훨씬 더 높은 농도(1000mg/L)가 발견되지만 대부분의 담수 공급에서 3~30mg/L 범위입니다.
- 전도도(Conductivity)
    - 순수한 물은 전류의 좋은 전도체가 아니라 좋은 절연체입니다. 이온 농도의 증가는 물의 전기 전도성을 향상시킵니다. 일반적으로 물에 용해된 고형물의 양은 전기 전도도를 결정합니다. 전기 전도도(EC)는 실제로 용액이 전류를 전달할 수 있도록 하는 용액의 이온 과정을 측정합니다. WHO 표준에 따르면 EC 값은 400μS/cm를 초과해서는 안 됩니다.
- 유기탄소(Organic_carbon)
    - 원수의 총유기탄소(TOC)는 썩어가는 천연유기물질(NOM)과 합성원에서 발생합니다. TOC는 순수한 물에 있는 유기 화합물의 총 탄소량을 측정한 것입니다. US EPA에 따르면 처리/음용수에서 TOC로 < 2 mg/L, 처리에 사용되는 원수에서 < 4 mg/Lit.
- 트리할로메탄(Trihalomethanes)
    - THM은 염소 처리된 물에서 발견될 수 있는 화학 물질입니다. 음용수의 THM 농도는 물의 유기 물질 수준, 물을 처리하는 데 필요한 염소의 양 및 처리되는 물의 온도에 따라 다릅니다. 최대 80ppm의 THM 수준은 식수에서 안전한 것으로 간주됩니다.
- 탁도(Turbidity)
    - 물의 탁도는 부유 상태에 존재하는 고형물의 양에 따라 달라집니다. 물의 발광 특성을 측정하는 척도로 콜로이드 물질에 대한 폐기물 배출 품질을 나타내는 데 사용됩니다. 원도 제넷캠퍼스에서 얻은 평균 탁도 값(0.98 NTU)은 WHO 권장 값인 5.00 NTU보다 낮습니다.
- 이식성(Potability)
    - 물이 사람이 섭취하기에 안전한지 여부를 나타냅니다. 여기서 1은 음용 가능, 0은 음용 불가를 의미합니다.

생각보다 정확도가 높지 않다는 걸 확인할 수 있었음

- <u> ***교차 검증 사용*** </u>
- K-fold와 Cross val을 사용하기로 함

```python
from sklearn.model_selection import KFold

def exec_kfold(model, title = "Default", folds=5):
    # 폴드 세트를 5개인 KFold객체를 생성, 폴드 수만큼 예측결과 저장을 위한  리스트 객체 생성.
    kfold = KFold(n_splits=folds)
    scores = []
    
    # KFold 교차 검증 수행. 
    for iter_count , (train_index, test_index) in enumerate(kfold.split(x_water_df)):
        # X_titanic_df 데이터에서 교차 검증별로 학습과 검증 데이터를 가리키는 index 생성
        x_train, x_test = x_water_df.values[train_index], x_water_df.values[test_index]
        y_train, y_test = y_water_df.values[train_index], y_water_df.values[test_index]
        
        # Classifier 학습, 예측, 정확도 계산 
        model.fit(x_train, y_train) 
        predictions = model.predict(x_test)
        accuracy = accuracy_score(y_test, predictions)
        scores.append(accuracy)
        # print("교차 검증 {0} 정확도: {1:.4f}".format(iter_count, accuracy))
    
    # 5개 fold에서의 평균 정확도 계산. 
    mean_score = np.mean(scores)
    print(title, "Mean Accuracy: {0:.4f}".format(mean_score), '\n') 
```

```python
# Naive Bayes
nb = GaussianNB()
exec_kfold(nb, "Naive Bayes", folds=5)

# Stochastic Gradient Descent
sgd = SGDClassifier() #(max_iter=5000, random_state=0)
exec_kfold(sgd, "Stochastic Gradient Descent", folds=5)

# KNN
knn = KNeighborsClassifier(n_neighbors=140)
exec_kfold(knn, "KNN", folds=5)

# Decission trees
tree = DecisionTreeClassifier()
exec_kfold(tree, "Decission trees", folds=5)

# Random Forest
rforest = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=0)
exec_kfold(rforest, "Random Forest", folds=5)

# Support Vector Machine
svm = SVC(decision_function_shape="ovo")
exec_kfold(svm, "Support Vector Machine", folds=5)

# Logistic Regression
lg = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')
exec_kfold(lg, "Logistic Regression", folds=5)

# Neural Nets
nn = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5000, 10), random_state=1)
exec_kfold(nn, "Neural Nets", folds=5)

# catboost
cbc = cb.CatBoostClassifier(verbose=0, eval_metric='Accuracy', loss_function='MultiClass')
exec_kfold(cbc, "catboost", folds=5)

# Cross Gradient Booster
xgb = XGBClassifier(n_estimators=1000, learning_rate=0.05)
exec_kfold(xgb, "Cross Gradient Booster", folds=5)

# Cross Gradient Booster (Random Forest)
xgbrf = XGBRFClassifier(objective= 'binary:logistic')
exec_kfold(xgbrf, "Cross Gradient Booster (Random Forest)", folds=5)
```

```python
from sklearn.model_selection import cross_val_score

def StratifiedKFold(model, title = "Default", folds=5):
    # 폴드 세트를 5개인 KFold객체를 생성, 폴드 수만큼 예측결과 저장을 위한  리스트 객체 생성.
    scores = cross_val_score(model, x_water_df , y_water_df , cv=5)

    # for iter_count,accuracy in enumerate(scores):
        # print("교차 검증 {0} 정확도: {1:.4f}".format(iter_count, accuracy))
        
    print(title, "Mean Accuracy: {0:.4f}".format(np.mean(scores)), '\n')
```

```python
# Naive Bayes
nb = GaussianNB()
StratifiedKFold(nb, "Naive Bayes", folds=5)

# Stochastic Gradient Descent
sgd = SGDClassifier() #(max_iter=5000, random_state=0)
StratifiedKFold(sgd, "Stochastic Gradient Descent", folds=5)

# KNN
knn = KNeighborsClassifier(n_neighbors=140)
StratifiedKFold(knn, "KNN", folds=5)

# Decission trees
tree = DecisionTreeClassifier()
StratifiedKFold(tree, "Decission trees", folds=5)

# Random Forest
rforest = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=0)
StratifiedKFold(rforest, "Random Forest", folds=5)

# Support Vector Machine
svm = SVC(decision_function_shape="ovo")
StratifiedKFold(svm, "Support Vector Machine", folds=5)

# Logistic Regression
lg = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')
StratifiedKFold(lg, "Logistic Regression", folds=5)

# Neural Nets
nn = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5000, 10), random_state=1)
StratifiedKFold(nn, "Neural Nets", folds=5)

# catboost
cbc = cb.CatBoostClassifier(verbose=0, eval_metric='Accuracy', loss_function='MultiClass')
StratifiedKFold(cbc, "catboost", folds=5)

# Cross Gradient Booster
xgb = XGBClassifier(n_estimators=1000, learning_rate=0.05)
StratifiedKFold(xgb, "Cross Gradient Booster", folds=5)

# Cross Gradient Booster (Random Forest)
xgbrf = XGBRFClassifier(objective= 'binary:logistic')
StratifiedKFold(xgbrf, "Cross Gradient Booster (Random Forest)", folds=5)
```

```python
def model_assess_(model, title = "Default"):
    model.fit(x_train, y_train)
    preds = model.predict(x_test)
    #print(confusion_matrix(y_test, preds))
    print('Accuracy', title, ':', round(accuracy_score(y_test, preds), 5))

def model_scaler_assess(model, title = "Default"):
    model.fit(x_scaled_train, y_scaled_train)
    preds = model.predict(x_scaled_test)
    #print(confusion_matrix(y_test, preds))
    print('StandardScaler Accuracy', title, ':', round(accuracy_score(y_scaled_test, preds), 5), '\n')
```

```python
x_water_df = water_df.drop('Potability', axis = 1)
y_water_df = water_df['Potability']

x_train, x_test, y_train, y_test = train_test_split(x_water_df,
                                                    y_water_df,
                                                    test_size=0.2,
                                                    random_state=45)

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_water_df)

x_scaled_train, x_scaled_test, y_scaled_train, y_scaled_test = train_test_split(x_scaled,
                                                                                y_water_df, 
                                                                                test_size = 0.2, 
                                                                                random_state = 45)

mm_scaler = MinMaxScaler()
x_mscaled = mm_scaler.fit_transform(x_water_df)
x_mscaled_train, x_mscaled_test, y_mscaled_train, y_mscaled_test = train_test_split(x_mscaled,
                                                                                    y_water_df, 
                                                                                    test_size = 0.2, 
                                                                                    random_state = 45)
```

```python
def model_assess(model, title = "Default"):
    model.fit(x_train, y_train)
    preds = model.predict(x_test)
    print(title)
    print('정확도 :', round(accuracy_score(y_test, preds), 5))

def model_kfold(model, folds=5):
    kfold = KFold(n_splits=folds)
    scores = []

    for iter_count , (train_index, test_index) in enumerate(kfold.split(x_water_df)):
        x_train, x_test = x_water_df.values[train_index], x_water_df.values[test_index]
        y_train, y_test = y_water_df.values[train_index], y_water_df.values[test_index]
        
        model.fit(x_train, y_train) 
        predictions = model.predict(x_test)
        accuracy = accuracy_score(y_test, predictions)
        scores.append(accuracy)
    
    mean_score = np.mean(scores)
    print("K-fold 평균 정확도: {0:.4f}".format(mean_score))

def StratifiedKFold(model, folds=5):
    scores = cross_val_score(model, x_water_df , y_water_df , cv=5)
    print("교차 검증 평균 정확도: {0:.4f}".format(np.mean(scores)))

def model_st_scaler_assess(model):
    model.fit(x_scaled_train, y_scaled_train)
    preds = model.predict(x_scaled_test)
    print('표준 정규분포 정확도:', round(accuracy_score(y_scaled_test, preds), 5))

def model_mm_scaler_assess(model):
    model.fit(x_mscaled_train, y_mscaled_train)
    preds = model.predict(x_mscaled_test)
    print('최소최대 정규분포 정확도:', round(accuracy_score(y_mscaled_test, preds), 5), '\n')
```

```python
def Accuracy_model(model, title):
    model_assess(model, title)
    model_kfold(model, folds=5)
    StratifiedKFold(model, folds=5)
    model_st_scaler_assess(model)
    model_mm_scaler_assess(model)
```

```python
# Naive Bayes
nb = GaussianNB()
Accuracy_model(nb, "Naive Bayes")

# Stochastic Gradient Descent
sgd = SGDClassifier() #(max_iter=5000, random_state=0)
Accuracy_model(sgd, "Stochastic Gradient Descent")

# KNN
knn = KNeighborsClassifier(n_neighbors=140)
Accuracy_model(knn, "KNN")

# Decission trees
tree = DecisionTreeClassifier()
Accuracy_model(tree, "Decission trees")

# Random Forest
rforest = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=0)
Accuracy_model(rforest, "Random Forest")

# Support Vector Machine
svm = SVC(decision_function_shape="ovo")
Accuracy_model(svm, "Support Vector Machine")

# Logistic Regression
lg = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')
Accuracy_model(lg, "Logistic Regression")

# Neural Nets
nn = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5000, 10), random_state=1)
Accuracy_model(nn, "Neural Nets")

# catboost
cbc = cb.CatBoostClassifier(verbose=0, eval_metric='Accuracy', loss_function='MultiClass')
Accuracy_model(cbc, "catboost")

# Cross Gradient Booster
xgb = XGBClassifier(n_estimators=1000, learning_rate=0.05)
Accuracy_model(xgb, "Cross Gradient Booster")

# Cross Gradient Booster (Random Forest)
xgbrf = XGBRFClassifier(objective= 'binary:logistic')
Accuracy_model(xgbrf, "Cross Gradient Booster (Random Forest)")
```
